```python
import pandas as pd
```

```python
data = pd.read_csv('/Users/macbookpro/Downloads/Fire_Incident_Dispatch_Data.csv')
```

```python
data.columns
```

```python
data['INCIDENT_DATE']= pd.to_datetime(data['INCIDENT_DATETIME']).dt.date
```

```python
data['year'] = pd.DatetimeIndex(data['INCIDENT_DATE']).year
```

```python
# datay = data[data['year'].isin(['2018','2019'])]
```

```python
datay = data
```

```python
datax = datay.groupby(['INCIDENT_DATE','INCIDENT_BOROUGH']).size().to_frame('incident_count').reset_index()
```

```python
datax
```

```python
datax.groupby(['INCIDENT_DATE','INCIDENT_BOROUGH']).sum()['incident_count'].unstack()
```

```python
import matplotlib.pyplot as plt
# plot data
fig, ax = plt.subplots(figsize=(15,7))
# use unstack()
datax.groupby(['INCIDENT_DATE','INCIDENT_BOROUGH']).sum()['incident_count'].unstack().plot(ax=ax)
# datax.groupby(['INCIDENT_DATE','INCIDENT_BOROUGH']).count()['incident_count'].plot(ax=ax)
```

```python
'''
from the graph, over the years, it can be viewed that, 
richmond state has least fire incidents not more than atleast 200 cases on average
with queens and manhattan recording more incidents every year

''' 
```

```python
datay['diff_minutes'] = pd.to_datetime(datay['INCIDENT_CLOSE_DATETIME']) - pd.to_datetime(datay['INCIDENT_DATETIME'])
```

```python
import numpy as np
```

```python
datay['diff_minutes'] = datay['diff_minutes']/np.timedelta64(1,'m')
```

```python
# data_2018 = data[datay['year'] == 2018]
```

```python
incident_data = data.groupby(['INCIDENT_BOROUGH','year'])['diff_minutes'].agg(['mean','count'])
```

```python
g, ax = plt.subplots(figsize=(15,7))
# use unstack()
incident_data.groupby(['year','INCIDENT_BOROUGH']).sum()['mean'].unstack().plot(ax=ax)
# datax.groupby(['INCIDENT_DATE','INCIDENT_BOROUGH']).count()['incident_count'].plot(ax=ax)
```
